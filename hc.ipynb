```python
# Necessary imports

import numpy
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
```

```python
# Make 2-dimensional blobs
data = make_blobs(n_samples = 400,
                 n_features = 2,
                 cluster_std = 1.2,
                 centers = 4,
                 random_state = 12)
```

```python
plt.scatter(data[0][:, 0], data[0][:, 1])
```

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean')
y_pred = hc.fit_predict(data[0])
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(data[1], y_pred)
```

```python
cm
```

```python
import seaborn as sns

sns.heatmap(cm)
```

```python
# Predicted labels
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1], hue = data[1])
```

```python
#Actual labels
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1], hue = y_pred)
```

```python
#silhouette_score
from sklearn.metrics import silhouette_score

print(silhouette_score(data[0], y_pred))
```
